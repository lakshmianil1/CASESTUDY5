# 3.Check whether all the zones generate the same amount of sales.
H0 (Null Hypothesis): all the zones generate the same amount of sales

H1 (Alternate Hypothesis): sales generated by all zones are different

```python
# first we need to import the libraries for one way ANOVA which is present in the scipy.stat model
import scipy.stats as stats
# find the f_statistic and p_value for one way ANOVA
f_statistic, p_value = stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

```python
# display the values of f_statistic and p_value
print('f_statistic=',round(f_statistic,5))
print('p_value=',round(p_value,5))
```

```python
# let's check the hypothesis analysis
if p_value<0.05:
    print('Reject the null hypothesis and we can say that all the zones generate the same amount of sales')
else:
    print('Accept the null hypothesis and we can say that  sales generated by all zones are different')
```
